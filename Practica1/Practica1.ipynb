```python
#Importación de la libreria pandas que se utilizara como el kernel principal de nuestra libreta, este tendra un alias 'pd'
import pandas as pd
```

```python
# En este apartado se hara una serie, esta es una estructura de datos de una sola dimensión, la instanciacion de este objeto se hace mediante 
# Crear serie 
ser1 = pd.Series([10,20,30,40])
# Mostrar en pantalla el otput de la serie
ser1
# El output mostrara los indices a la izquierda y los valores a la derecha
```

```python
# En las series se pueden representar diferentes tipos de datos
ser2 = pd.Series([[10,20], \
                  [30,40,'series'],\
                    [50,55],
                    {'Name':'Tess','Org':'Packt'}])
ser2
```

```python
#Los dataframes son estructuras de datos bidimensionales comunmente usados para manipular y analizar 
# datos originarios de diferentes fuentes. Esta tiene filas y columnas las cuales se inicializan usando el constructor DataFrame().
#Tenemos de ejemplo la siguiente lista convertida a un DataFrame de una dimension 
#Crear un DataFrame usando el constructor
df=pd.DataFrame([30,50,20]) 
#Mostrar en pantalla el DataFrame
df
```

```python
#La forma de un DataFrame, es decir su tupla con sus dimensiones se visualiza usando este atributo 
df.shape
#El output muestra el numero de filas y columnas respectivamente 
```

```python
# Las columnas tienen nombres asisgnados en lugar de los indices que estan en las filas. Para mostrar los nombres de las columnas
#se utiliza el atributo columns que al igual que el atributo index para ver los indices muestran en que número inician, en que 
#nombre paran y de cuanto es el avance
df.columns 
```

```python
#El nombre de los indices se pueden mostrar con el siguiente atributo 
df.index
```

```python
#Para mostrar los nombres de las filas y columnas se hace lo siguiente 
print("Estas son el nombre de las columnas",list(df.columns))
print ("Estos son los nombres de las filas o indices", list(df.index))
```

```python
#Las columnas y los indices se pueden renombrar asignandoles lista de valores 
#Lista de valores para las columnas
df.columns=['V1']
df
```

```python
#Lista de valores para los indices 
df.index=['R1','R2','R3']
df
```

```python
#Para crear un DataFrame con varias columnas, se necesita una lista anidada de listas 
df1=pd.DataFrame(([10,15,20],[100,200,300]))
print ("Tupla de la nueva data frame",df1.shape)
df1
```

```python
#Este comando es para asignar los nombres de las columnas y el nombre de las filas mientras se crea el DataFrame 
df1=pd.DataFrame([[10,15,20],[100,200,300]],\
                 columns=['V1','V2','V3'],\
                    index=['R1','R2'])
df1
```
