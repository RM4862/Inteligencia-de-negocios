```python
#Para convertir un dataset a documento se usa el comando df.to_csv(outpath).El df es el nombre del dataframe 
#y el outpath es el parametro para definir la ruta donde los datos son escritos
import pandas as pd 
filename='student-por.csv'
#Comando para leer el dataset donde se guardara en una variable llamada studentdata
studentData = pd.read_csv(filename,delimiter=';')
#leemos las primeras 5 lineas del documento 
studentData.head()

```

```python
#Creamos un nuevo dataframe con las primeras 5 filas
studentSmall=studentData.head()
outpath = 'studentSmall.csv'
#Comando para escribir este nuevo dataset al disco 
studentSmall.to_csv(outpath)
```
